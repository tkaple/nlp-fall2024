```python
import json
occupation = "businessperson"
total_parts = 186

wikidata_person = []
for i in range(1,total_parts):
    with open('./Data/English_Labelled_Wikidata_'+occupation+'/en_labelled_part'+str(i)+'.json') as f:
        new_list = json.load(f)
        wikidata_person.append(new_list)
```

```python
properties_list = {}

count = 0
for batch in wikidata_person:
    for qid in batch.keys():
        count+=1
        for prop in batch[qid].keys():
            if prop not in properties_list:
                properties_list[prop] = 1
            else:
                properties_list[prop] +=1
```

```python
len(properties_list)
```

```python
id_props = ['official website', 'Commons Creator page', 'Commons gallery', 'ISNI', 'signature', 'subreddit', 'U.S. National Archives Identifier', 'Libris-URI', 'Wolfram Language entity code', 'pronunciation audio', 'Europeana entity', 'Munzinger Sport number', 'described at URL', 'image of grave', 'plaque image', 'maintained by WikiProject', 'coat of arms image', 'name in kana']

for prop in properties_list:
    if "ID" in prop:
        id_props.append(prop)
    if "username" in prop:
        id_props.append(prop)
```

```python
for prop in id_props:
    del properties_list[prop]
```

```python
len(properties_list)
```

```python
low_freq_props = []

for prop in properties_list.keys():
    if properties_list[prop] <= int(count*0.01):
        low_freq_props.append(prop)

for prop in low_freq_props:
    del properties_list[prop]
```
